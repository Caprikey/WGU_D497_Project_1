```python
def get_session_token(session_url):

    try:
        get_session_response = requests.get(session_url)
        get_session_response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return None 
        
    try: 
        session_soup = BeautifulSoup(get_session_response.text, "html.parser")
    except Exception as e:
        print(f"BeautifulSoup failed to parse the response: {e}")
        return None

    try:
        nonce_element = session_soup.find("input", {"id": "wdtNonceFrontendServerSide_1"})
        if nonce_element is not None:
            #wdt_nonce = nonce_element["value"]
            wdt_nonce = nonce_element.get("value")
        else:
            print("Nonce Element not found.")
            return None

    except Exception as e: 
        print(f"Error while extracting wdtNonce: {e}")
        return None
            
    return wdt_nonce
```

```python
def get_api_response(api_url, api_payload=None, api_headers=None, print_response=True):
    
    """
    Sends a POST request to the specified API URL with the given payload and headers.
    
    Args:
        api_url (str): The API endpoint URL.
        api_payload (dict, optional): The data to be sent in the POST request. Defaults to None.
        api_headers (dict, optional): The headers to be sent with the request. Defaults to None.
        print_response (bool, optional): Whether to print the response. Defaults to True.
        
    Returns:
        dict | None: The parsed JSON response if successful, or None if an error occurs.
    """

    
    #api_response = api_response = requests.post(api_url, data=api_payload, headers=api_headers)

    try:
        api_response = requests.post(api_url, data=api_payload, headers=api_headers)
        api_response.raise_for_status()

        if api_response.status_code == 200:
            try:
                api_response_data = api_response.json()
                if print_response:
                    print("JSON Response:", api_response_data)
                return api_response_data
            except ValueError:
                if print_response:
                    print("Non-JSON Response:", api_response.text[:500])
                    return None
        else:
            print("Error:", api_response.status_code, api_response.text)

    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return None
```

```python
def get_total_records(api_response_data):
    api_total_records = api_response_data['recordsTotal']
    return api_total_records
```

```python
def get_total_loops(api_total_records):
    loop_cycles = int(api_total_records) // 1000

    if (loop_cycles % 1000) < 1000:
        loop_cycles += 1

        total_loop_cycles = loop_cycles

    return total_loop_cycles
```

```python
def extract_api_data(api_response_data):

    global extracted_data
    
    api_table_data = api_response_data['data']

    for row in api_table_data:

        
        #print(row)
        #print()
    
        #print(row)
    
        extracted_data_row = []
        
        # Extracting Href Value Assigning First Index item to a working/staging variable
        link_text_raw = row[0]
    
        # Searching Link Text Raw with RE to locate the href value 
        link_text_extraction = re.search(r"href='(.*?)'", link_text_raw)
    
    
        record_id_extraction = re.search(r"sighting\/\?id=(.*?)(')", link_text_raw)
        record_id = record_id_extraction.group(1)
        
        extracted_data_row.append(record_id)
    
        #extracted_data_row.append(occurred)
        extracted_data_row.append(row[1])
    
        #extracted_data_row.append(city)
        extracted_data_row.append(row[2])
        
        #extracted_data_row.append(state)
        extracted_data_row.append(row[3])
        
        #extracted_data_row.append(country)
        extracted_data_row.append(row[4])
        
        #extracted_data_row.append(shape)
        extracted_data_row.append(row[5])
    
        #extracted_data_row.append(summary)
        extracted_data_row.append(row[6])
        
        #extracted_data_row.append(reported)
        extracted_data_row.append(row[7])
        
        #extracted_data_row.append(media)
        #extracted_data_row.append(row[8])
        extracted_data_row.append('N' if row[8] is None else row[8])
    
        
        #extracted_data_row.append(explanation)
        extracted_data_row.append(row[9])
    
        # Assigning the final value with the results from the first group from the RE Extraction 
        if link_text_extraction:
            link_text_final = base_report_link + link_text_extraction.group(1)
        else:
            # Assigning link text to main thread 
            link_text_final = base_report_link + "/subndx/?id=all"
    
        extracted_data_row.append(link_text_final)
    
    
        
        extracted_data.append(extracted_data_row)



    return extracted_data
```

```python
def export_results_to_csv(data, export_start_index, export_end_index):
    
    nuforc_data = pd.DataFrame(data, columns=column_names)

    base_file_name = "raw_ufo_data_extraction_results_"

    file_name = base_file_name + str(export_start_index+1) + "_to_" + str(export_end_index)
    
    print(f"Exporting NUFORC Data - Results: {export_start_index} to {export_end_index} ")
    
    nuforc_data.to_csv(config.global_raw_ufo_data_folder_path + "/" + file_name + ".csv", index=False)

    data.clear()
    
    print(f"Export Completed")
```

```python
def main():

    global draw_index
    global start_index 
    
    # Step 1: Obtain a session token (nonce) for the Web Data Transfer (WDT) process
    get_wdt_nonce = get_session_token(session_url)
    
    # Store the nonce in a variable for reuse
    wdt_nonce = get_wdt_nonce
    
    # Step 2: Generate the initial API payload with necessary parameters
    initial_api_payload = generate_initial_api_payload(wdt_nonce, draw_index, start_index, results_length)
    
    # Step 3: Fetch data from the API using the initial payload
    api_response_data = get_api_response(api_url, api_payload=initial_api_payload, api_headers=api_headers, print_response=False)
    
    # Step 4: Retrieve the total number of records from the API response
    api_total_records = get_total_records(api_response_data)
    
    # Step 5: Calculate the total number of loop cycles needed to process all records
    total_loop_cycles = get_total_loops(api_total_records)
    
    # Initialize loop cycle counter
    loop_cycle = 0
    
    # Flag for controlling data export at the end of a loop
    export_last_loop_flag = False
    
    # Counter for the number of times data is exported within a loop
    export_loop_counter = 0
    
    # Set the limit for the number of times data will be exported in a loop
    export_loop_limit = 10
    
    # Starting index for exporting data
    export_start_index = 0
    
    # Step 6: Begin looping through API data in cycles
    while loop_cycle <= total_loop_cycles:
    
        # Extract data from the current API response
        extract_api_data(api_response_data)
            
        # Increment the draw_index for the next request
        draw_index += 1
        
        # Increment the start index for the next request
        start_index += int(results_length)
        
        # Calculate the end index for the data export in this cycle
        export_end_index = export_start_index + (results_length * (export_loop_counter + 1))
    
        # Step 7: Generate the API payload for the next set of paginated data
        pagination_api_payload = generate_pagination_api_payload(wdt_nonce, draw_index, start_index, results_length)
            
        # Fetch the next API response using the paginated payload
        api_response_data = get_api_response(api_url, api_payload=pagination_api_payload, api_headers=api_headers, print_response=False)
        
        # Increment the loop counter for data exports
        export_loop_counter += 1
    
        # Step 8: Check if the number of exports has reached the defined limit
        if export_loop_counter == export_loop_limit:
    
            # Export the extracted data to a CSV file
            export_results_to_csv(extracted_data, export_start_index , export_end_index )
            
            # Reset the export counter for the next batch
            export_loop_counter = 0
                    
            # Update the start index for the next batch of data export
            export_start_index  = export_end_index 
    
        # Step 9: Check if we are at the final loop cycle
        if loop_cycle == total_loop_cycles:
            
            # Export the remaining extracted data to CSV
            export_results_to_csv(extracted_data, export_start_index , export_end_index)
        
        # Increment the loop cycle counter
        loop_cycle +=  1
```

```python
if __name__ == "__main__":
    main()
```
