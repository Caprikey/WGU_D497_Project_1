```python
def find_raw_files(year:str):

    year_selected =  year

    # Used By Glob
    file_name_pattern = "raw_cdc_data_*.txt"

    # Used By Matching Indices
    re_pattern = r"raw_cdc_data_" + year + r"\.txt"
    
    #re.search((^.+(CDC_Natality_Birth_Data.)+(Raw_CDC_Natality_Birth_Data_)+)(\d+)+(.txt)

    matched_files = glob.glob(config.global_raw_cdc_data_folder_path + "/" + file_name_pattern)

    if matched_files:
    
        matched_files.sort()

        matching_indices = [index for index, file in enumerate(matched_files) if re.search(re_pattern, file)]
        
        if matching_indices:
            #print("Found")
            cdc_data_file = matching_indices[0]
            cdc_data_file_path = matched_files[cdc_data_file]
            return cdc_data_file_path
        else:
            #print("Not Found")
            return []
    
        
        return cdc_data_file
        
    else:
        print(f"No Files Found That Match the File Name Pattern {file_name_pattern}")
        return []

```

```python
def get_footer(file_path, rows_to_skip):

    with open(file_path, 'r') as f:
        lines = f.readlines()
        return lines[-rows_to_skip:]
```

```python
def save_footer(lines, year):
    
    export_folder_path = config.global_cdc_footer_folder_path
    #print(cdc_footers_folder)
    export_footer_file_name = "CDC_Footer_Data_" + str(year) + ".txt"

    export_file_footer_path = Path(export_folder_path + "/" + export_footer_file_name)
    
    #print(export_file_footer_path)
    f = open(export_file_footer_path, "w")
    f.writelines(lines)
    f.close()
```

## Data Cleaning
### Import Downloaded Year Data and Split into Two files, Data file and Footer file

```python
for year in range(year_start, year_end+1):
    
    file_path = find_raw_files(str(year))

    import_dictionary[year] = {
        "file_path" : file_path, 
        "footer_start_index" : None,
        "length_of_file": None,
        "rows_to_skip": None,
    }
    
    #cdc_data_file = pd.read_csv(file_path, sep="\t")

    #print(f"File for {year} found at {file_path}")

    # Step 1: Read the file to detect the position of the first "---"
    raw_df = pd.read_csv(file_path, engine='python', sep="\t")

    # Step 2: Find the index where the first occurrence of '---' appears in the first column
    footer_start_index = raw_df[raw_df.iloc[:, 0].astype(str).eq('---')].index.min()

    import_dictionary[year].update({"footer_start_index": footer_start_index})

    total_length = len(raw_df)

    import_dictionary[year].update({"length_of_file": total_length})
    
    # Step 3: Calculate the number of rows to skip (total rows - footer start position)
    #rows_to_skip = len(raw_df) - footer_start_index
    rows_to_skip = total_length - footer_start_index
    import_dictionary[year].update({"rows_to_skip": rows_to_skip})
    
    #print(f"Footer starts at: {footer_start_index}, footer ends at {len(raw_df)}, and is {rows_to_skip} rows long\n")
    #print(f"Footer starts at: {footer_start_index}, footer ends at {total_length}, and is {rows_to_skip} rows long\n")
    
    total_rows += footer_start_index
    print(f"{year} should add {(total_length-rows_to_skip)} to the total rows counter, new value is: {total_rows}")
    
    if year <= 2002:
        # Step 4a: Re-read the file, skipping footer rows, uses "dtypes_old" for handling the older years structure/data differences
        cdc_data_partial_df = pd.read_csv(file_path, skipfooter=rows_to_skip, engine='python', sep="\t", dtype=dtypes_old)
        
    else:
        # Step 4b: Re-read the file, skipping footer rows, uses "dtypes_new" for handling the newer years structure/data differences 
        cdc_data_partial_df = pd.read_csv(file_path, skipfooter=rows_to_skip, engine='python', sep="\t", dtype=dtypes_new)
        
    # Step 5 : Assigning Variables for Path and File Name Creationg
    export_folder_path = config.global_processed_cdc_data_folder_path
    export_file_name = "processed_cdc_data_" + str(year) + "_footer_removed.csv"

    # Step 6 :: Creates Export File Path For CSV Export
    export_file_path = Path(export_folder_path + "/" + export_file_name)
    #print(export_file_path)


    footer_lines = get_footer(file_path, rows_to_skip)
    #print(footer_lines)

    save_footer(footer_lines, year)
    
    # Step 7 :: Exports The Partial Dataframe or the imported year's data without the footer to CSV file
    cdc_data_partial_df.to_csv(export_file_path, sep=",")
        
    # Step 8 :: Combined CDC Data and concat into main
    cdc_data_main_footless_df = pd.concat([cdc_data_main_footless_df, cdc_data_partial_df], axis=0, ignore_index=True)

    # Step 9 :: Delete Partial CDC Data Dataframe to free up memory before looping again
    del cdc_data_partial_df

    # Step 10 :: Delete Raw Dataframe to free up memory before looping again 
    del raw_df
```
