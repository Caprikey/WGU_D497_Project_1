```python
# function strips any new lines, white spaces, etc from the columns and keep track of any changes made 

def strip_and_track(text):

    if not isinstance(text, str):
        return text

    def strip_value(match):

        global stripping_needed
        global strip_match_counter
        
        stripping_needed = True
        strip_match_counter['count'] += 1
        return ''  #

    match_pattern = re.compile(r'\r+|\n+|\t+')
    
    return match_pattern.sub(strip_value, text)
```

```python

# This cell files all the original web scraped files from the data extraion and captures their paths into a list variable

file_name_pattern = "raw_ufo_data_extraction_results_*.csv"


matched_files = glob.glob(config.global_raw_ufo_data_folder_path + "/" + file_name_pattern )

if matched_files:
    matched_files.sort()
    #print(matched_files)

else:
    print(f"No Files Found That Match the File Name Pattern {file_name_pattern}")

```

```python
# This cell loops through finding all files that where found to match the search pattern in the cell above. 
# On each matched file, the loop strips and cleans alll column values to remove any new lines, white space, etc. 
# Each file's changes are kept track in the cleaning dictionary

cleaning_dictionary = {}

for file in matched_files:
    
    file_name_match_pattern = re.search(r"(\d+_to_\d+)", str(file))

    file_name_result = file_name_match_pattern.group(1)
    
    #print(file_name_index_match_pattern)

    cleaning_dictionary[file_name_result] = {
        "import_file" : file, 
        "import_file_total_before_strip" : None,
        "import_file_total_after_strip": None,
        "stripping_needed": False,
        "strip_count": 0,
        "export_file": None,
        "export_file_total_after_csv_export": None,
    }

    partial_ufo_data = pd.read_csv(file)

    import_file_total_before_strip = len(partial_ufo_data)

    cleaning_dictionary[file_name_result].update({"import_file_total_before_strip":import_file_total_before_strip})

    # This works but does not track changes made
    #partial_ufo_data.replace(r'\r+|\n+|\t+','', regex=True, inplace=True)

    # This should apply to the whole dataframe but fails if something other than a string is found
    #partial_ufo_data = partial_ufo_data.applymap(strip_and_track)
    
    for col in partial_ufo_data.columns:
        
        if partial_ufo_data[col].dtype == 'object':
            
            partial_ufo_data[col] = partial_ufo_data[col].apply(strip_and_track)
            

    cleaning_dictionary[file_name_result].update({"stripping_needed": stripping_needed, "strip_count": strip_match_counter['count']})

    import_file_total_after_strip = len(partial_ufo_data)

    cleaning_dictionary[file_name_result].update({"import_file_total_after_strip":import_file_total_after_strip})

    export_folder_path = config.global_processed_ufo_data_folder_path

    export_file_name = "processed_ufo_data_extraction_results_" + file_name_result + "_stripped.csv"

    # Step 6 :: Creates Export File Path For CSV Export
    export_file_path = Path(config.global_processed_ufo_data_folder_path + "/" + export_file_name)
    #print(export_file_path)
    
    partial_ufo_data.to_csv(export_file_path, sep=",")

    cleaning_dictionary[file_name_result].update({"export_file":export_file_path})

    #raw_main_ufo_data_df = pd.concat([raw_main_ufo_data_df, partial_ufo_data], ignore_index=True)

    export_total_check_df = pd.read_csv(export_file_path)

    export_file_total = len(export_total_check_df)

    cleaning_dictionary[file_name_result].update({"export_file_total_after_csv_export": export_file_total})

    del partial_ufo_data

    del export_total_check_df
```

```python
# In this cell, we will loop through all the matched files above to split the data into two seperate files. 
# One file will contain all the columns of interest to this analysis.
# The second file will contain all the columns to be dropped from this analysis.
# However, these will be saved in case they are needed in the future. 
# I will include the report id on both files for each entry so that they may be matched easily later. 
# Like the last loop, this one will also keep track of the files in the cleaning dictionary 


for file in matched_files:
    
    file_name_match_pattern = re.search(r"(\d+_to_\d+)", str(file))

    file_name_result = file_name_match_pattern.group(1)
    

    cleaning_dictionary[file_name_result] = {
        "import_file" : file, 
        "import_file_total_before_drop" : None,
        "import_file_total_after_drop": None,
        "export_file_dropped": None,
        "export_file_dropped_total_after_csv_export": None,
        "export_file_kept": None,
        "export_file_kept_total_after_csv_export": None
    }


    raw_import_df = pd.read_csv(file, sep=",")

    raw_partial_ufo_data_df = raw_import_df.copy()

    del raw_import_df

    
    import_file_total_before_drop = len(raw_partial_ufo_data_df)

    cleaning_dictionary[file_name_result].update({"import_file_total_before_drop":import_file_total_before_drop})


    partial_ufo_data_dropped_columns_df = raw_partial_ufo_data_df[['report_id', 'shape','summary','explanation','has_media', 'report_link']]
    
    partial_ufo_data_kept_columns_df = raw_partial_ufo_data_df[['report_id', 'occurred_on', 'city', 'state','country', 'reported_on']]

    
    del raw_partial_ufo_data_df

    
    import_file_total_after_drop = len(partial_ufo_data_kept_columns_df)

    cleaning_dictionary[file_name_result].update({"import_file_total_after_drop":import_file_total_after_drop})


    
    export_folder_path = config.global_processed_ufo_data_folder_path

    export_dropped_column_file_name = "processed_ufo_data_extraction_results_" + file_name_result + "_column_drop.csv"
    
    export_dropped_column_file_path = Path(config.global_processed_ufo_data_folder_path + "/" + export_dropped_column_file_name)

    cleaning_dictionary[file_name_result].update({"export_file_dropped": str(export_dropped_column_file_path)}) 
    
    
    partial_ufo_data_dropped_columns_df.to_csv(export_dropped_column_file_path, sep=",")


    del partial_ufo_data_dropped_columns_df

    
    
    export_kept_column_file_name = "processed_ufo_data_extraction_results_" + file_name_result + "_column_kept.csv"
    
    export_kept_column_file_path = Path(config.global_processed_ufo_data_folder_path + "/" + export_kept_column_file_name)


    cleaning_dictionary[file_name_result].update({"export_file_kept": str(export_kept_column_file_path)})
    
    partial_ufo_data_kept_columns_df.to_csv(export_kept_column_file_path, sep=",")
 

    del partial_ufo_data_kept_columns_df
    

    export_dropped_total_check_df = pd.read_csv(export_dropped_column_file_path)

    export_dropped_file_total = len(export_dropped_total_check_df)

    cleaning_dictionary[file_name_result].update({"export_file_dropped_total_after_csv_export": export_dropped_file_total})

    del export_dropped_total_check_df

    export_kept_total_check_df = pd.read_csv(export_kept_column_file_path)

    export_kept_file_total = len(export_kept_total_check_df)
    
    cleaning_dictionary[file_name_result].update({"export_file_kept_total_after_csv_export": export_kept_file_total})

    del export_kept_total_check_df

    
```

#### Second pass has now been completed. Now it's time to do our third pass of the city data.

For the third pass, we will be using the RapidFuzz library to perform fuzzy query searching.

```python
# We will now do a line by line fuzzy query on each records. THIS PROCESS IS LONG
for idx, row in ufo_data_main_df_cities_merge_pass_three.iterrows():
    # Find the best fuzzy match for the city name in city_county_lookup
    match = process.extractOne(row['city'], county_city_lookup['city_lookup'])
    
    # If the match is above a threshold, update city_code and county_code
    if match and match[1] > 80:  # Threshold of 80 for match quality
        matched_row = county_city_lookup[county_city_lookup['city_lookup'] == match[0]].iloc[0]
        ufo_data_main_df_cities_merge_pass_three.at[idx, 'city_fipcode'] = matched_row['city_fipcode']
        ufo_data_main_df_cities_merge_pass_three.at[idx, 'county_fipcode'] = matched_row['county_fipcode']
```
