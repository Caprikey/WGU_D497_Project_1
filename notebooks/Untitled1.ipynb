```python
county_count_dict = cdc_data_main_df[cdc_data_main_df['county_fipcode'] != "C999"].groupby('state_fipcode')['county_fipcode'].nunique().to_dict()
```

```python
unspecified_counties = cdc_data_main_df[cdc_data_main_df['county_fipcode'] == "C999"]
```

```python
county_count_dict
```

```python
cdc_data_main_df[cdc_data_main_df['state_fipcode'] == "12086"]
```

```python
distribute_births = []
```

```python
for index, row in unspecified_counties.iterrows():

    state = row['state_fipcode']
    total_counties = county_count_dict.get(state, 1)
    total_births = row['births']

    per_county = total_births // total_counties
    remainder = total_births % total_counties 

    counties_df = cdc_data_main_df[(cdc_data_main_df['state_fipcode'] == state) & (cdc_data_main_df['county_fipcode'] != "C999")].copy()

    counties_df['extra_births'] = per_county

    if remainder > 0:
        remainder_index = np.random.choice(counties_df.index, size=remainder, replace=False)
        counties_df.loc[remainder_index, 'extra_births'] += 1

    distribute_births.append(counties_df)

distributed_births_df = pd.concat(distribute_births)

cdc_data_main_df = cdc_data_main_df[cdc_data_main_df['county_fipcode'] != "C999"]

cdc_data_main_df = pd.merge(cdc_data_main_df, distributed_births_df[['fips_five', 'extra_births']], on='fips_five', how="left")

cdc_data_main_df['births'] += cdc_data_main_df['extra_births'].fillna(0).astype(int)

cdc_data_main_df = cdc_data_main_df.drop(columns='extra_births')

```
