```python
def setup_working_directories():

    global raw_cdc_natality_birth_data_folder_path
    global cdc_footers_folder
    #fm.get_folder_path("raw_CDC_Natality_Birth_Data")

    raw_cdc_natality_birth_data_folder_path = fm.get_folder_path("raw_CDC_Natality_Birth_Data")
    
    cdc_footers_folder = fm.get_folder_path("cdc_footer_folder")
    
    #return raw_CDC_Natality_Birth_Data_Folder_Path, cdc_footers_folder
```

```python
def find_raw_files(year:str):

    year_selected =  year

    # Used By Glob
    file_name_pattern = "Raw_CDC_Natality_Birth_Data_*.txt"

    # Used By Matching Indices
    re_pattern = r"Raw_CDC_Natality_Birth_Data_" + year + r"\.txt"
    
    #re.search((^.+(CDC_Natality_Birth_Data.)+(Raw_CDC_Natality_Birth_Data_)+)(\d+)+(.txt)

    matched_files = glob.glob(raw_cdc_natality_birth_data_folder_path + "/" + file_name_pattern)

    if matched_files:
    
        matched_files.sort()

        matching_indices = [index for index, file in enumerate(matched_files) if re.search(re_pattern, file)]
        
        if matching_indices:
            #print("Found")
            cdc_data_file = matching_indices[0]
            cdc_data_file_path = matched_files[cdc_data_file]
            return cdc_data_file_path
        else:
            #print("Not Found")
            return []
    
        
        return cdc_data_file
        
    else:
        print(f"No Files Found That Match the File Name Pattern {file_name_pattern}")
        return []

```

```python
def find_totals_files(year:str):

    year_selected =  year

    # Used By Glob
    totals_file_name_pattern = "raw_CDC_Natality_Birth_Data_*_totals_rows.csv"
    #file_name_pattern = "CDC_Natality-Birth_Data_*_total_rows_removed.pkl"

    # Used By Matching Indices
    totals_re_pattern = r"raw_CDC_Natality_Birth_Data_" + year + r"\_totals_rows.csv"
    #re_pattern = r"CDC_Natality-Birth_Data_" + year + r"*_total_rows_removed.pkl"
    
    #re.search((^.+(CDC_Natality_Birth_Data.)+(Raw_CDC_Natality_Birth_Data-)+)(\d+)+(.txt)

    totals_matched_files = glob.glob(raw_cdc_natality_birth_data_folder_path + "/" + totals_file_name_pattern)

    if totals_matched_files:
    
        totals_matched_files.sort()

        totals_matching_indices = [index for index, file in enumerate(totals_matched_files) if re.search(totals_re_pattern, file)]
        
        if totals_matching_indices:
            #print("Found")
            totals_cdc_data_file = totals_matching_indices[0]
            totals_cdc_data_file_path = totals_matched_files[totals_cdc_data_file]
            return totals_cdc_data_file_path
        else:
            #print("Not Found")
            return []
    
        
        return totals_cdc_data_file
        
    else:
        print(f"No Files Found That Match the File Name Pattern {totals_file_name_pattern}")
        return []
```

```python
def get_footer(file_path, rows_to_skip):

    with open(file_path, 'r') as f:
        lines = f.readlines()
        return lines[-rows_to_skip:]
```

```python
def save_footer(lines, year):
    
    export_folder_path = cdc_footers_folder
    #print(cdc_footers_folder)
    export_footer_file_name = "CDC_Natality_Birth_Data_" + str(year) + "_footer.txt"

    export_file_footer_path = Path(export_folder_path + "/" + export_footer_file_name)
    
    #print(export_file_footer_path)
    f = open(export_file_footer_path, "w")
    f.writelines(lines)
    f.close()
```

## Data Cleaning
### Import Downloaded Year Data and Split into Two files, Data file and Footer file

```python
for year in range(year_start, year_end+1):
    
    file_path = find_raw_files(str(year))

    import_dictionary[year] = {
        "file_path" : file_path, 
        "footer_start_index" : None,
        "length_of_file": None,
        "rows_to_skip": None,
    }
    
    #cdc_data_file = pd.read_csv(file_path, sep="\t")

    #print(f"File for {year} found at {file_path}")

    # Step 1: Read the file to detect the position of the first "---"
    raw_df = pd.read_csv(file_path, engine='python', sep="\t")

    # Step 2: Find the index where the first occurrence of '---' appears in the first column
    footer_start_index = raw_df[raw_df.iloc[:, 0].astype(str).eq('---')].index.min()

    import_dictionary[year].update({"footer_start_index": footer_start_index})

    total_length = len(raw_df)

    import_dictionary[year].update({"length_of_file": total_length})
    
    # Step 3: Calculate the number of rows to skip (total rows - footer start position)
    #rows_to_skip = len(raw_df) - footer_start_index
    rows_to_skip = total_length - footer_start_index
    import_dictionary[year].update({"rows_to_skip": rows_to_skip})
    
    #print(f"Footer starts at: {footer_start_index}, footer ends at {len(raw_df)}, and is {rows_to_skip} rows long\n")
    #print(f"Footer starts at: {footer_start_index}, footer ends at {total_length}, and is {rows_to_skip} rows long\n")
    
    total_rows += footer_start_index
    print(f"{year} should add {(total_length-rows_to_skip)} to the total rows counter, new value is: {total_rows}")
    
    if year <= 2002:
        # Step 4a: Re-read the file, skipping footer rows, uses "dtypes_old" for handling the older years structure/data differences
        CDC_Natality_Birth_Data_Partial_df = pd.read_csv(file_path, skipfooter=rows_to_skip, engine='python', sep="\t", dtype=dtypes_old)
        
    else:
        # Step 4b: Re-read the file, skipping footer rows, uses "dtypes_new" for handling the newer years structure/data differences 
        CDC_Natality_Birth_Data_Partial_df = pd.read_csv(file_path, skipfooter=rows_to_skip, engine='python', sep="\t", dtype=dtypes_new)
        
    # Step 5 : Assigning Variables for Path and File Name Creationg
    export_folder_path = raw_cdc_natality_birth_data_folder_path
    export_file_name = "raw_CDC_Natality_Birth_Data_" + str(year) + "_footer_removed.csv"

    # Step 6 :: Creates Export File Path For CSV Export
    export_file_path = Path(export_folder_path + "/" + export_file_name)
    #print(export_file_path)


    footer_lines = get_footer(file_path, rows_to_skip)
    #print(footer_lines)

    save_footer(footer_lines, year)
    
    # Step 7 :: Exports The Partial Dataframe or the imported year's data without the footer to CSV file
    CDC_Natality_Birth_Data_Partial_df.to_csv(export_file_path, sep=",")
        
    # Step 8 :: Combined CDC Data and concat into main
    CDC_Natality_Birth_Data_Main_Footless_df = pd.concat([CDC_Natality_Birth_Data_Main_Footless_df,CDC_Natality_Birth_Data_Partial_df], axis=0, ignore_index=True)

    # Step 9 :: Delete Partial CDC Data Dataframe to free up memory before looping again
    del CDC_Natality_Birth_Data_Partial_df

    # Step 10 :: Delete Raw Dataframe to free up memory before looping again 
    del raw_df
```

```python
for year in range(year_start, year_end+1):

    feetless_file_path = find_feetless_files(str(year))

    #print(feetless_file_path)

    #import_dictionary[year] = {
    #    "file_path" : file_path, 
    #    "footer_start_index" : None,
    #    "length_of_file": None,
    #    "rows_to_skip": None,
    #}

    raw_cdc_natality_data_partial_df = pd.DataFrame()
    #cdc_data_file = pd.read_csv(file_path, sep="\t")

    #print(f"File for {year} found at {file_path}")

    # Step 1: Read the file to detect the position of the first "---"
    #raw_df = pd.read_csv(file_path, engine='python', sep="\t")

    if year <= 2002:
        # Step 4a: Re-read the file, skipping footer rows, uses "dtypes_old" for handling the older years structure/data differences
        raw_cdc_natality_data_partial_df = pd.read_csv(feetless_file_path, engine='python', dtype=dtypes_old)
        
    else:
        # Step 4b: Re-read the file, skipping footer rows, uses "dtypes_new" for handling the newer years structure/data differences 
        raw_cdc_natality_data_partial_df = pd.read_csv(feetless_file_path, engine='python', dtype=dtypes_new)


    # Get Length of Dataframe
    total_length = len(raw_cdc_natality_data_partial_df)

    # Index Match Rows where the "Notes" column has a value of "Total"
    totals_rows_df = raw_cdc_natality_data_partial_df.loc[raw_cdc_natality_data_partial_df['Notes'] == "Total"]

    # Index Match Rows where the "Notes" column does NOT have a value of "Total" 
    #totals_removed_df = totals_rows_df.loc[~totals_rows_df['Notes'] == "Total"]
    totals_removed_df = raw_cdc_natality_data_partial_df.loc[raw_cdc_natality_data_partial_df['Notes'] != "Total"]

    #print(totals_removed_df.shape[0])

    # Get Length of Non-Totals Dataframe
    new_total_length = len(totals_removed_df)

    # Get a Total value of rows to be removed 
    total_rows_removed = total_length - new_total_length 


    # Export folder Path
    #export_folder_path = raw_cdc_natality_birth_data_folder_path
    
    # Export file Name :: Totals Rows DF Export 
    #export_totals_rows_removed_file_name_csv = "CDC_Natality_Birth_Data_" + str(year) + "_totals_rows_removed.csv"
    #export_totals_rows_removed_file_name_pickle = "CDC_Natality_Birth_Data_" + str(year) + "_totals_rows_removed.pkl"

    # Export file Path :: Totals Rows DF Export 
    #export_totals_rows_removed_file_path_csv = Path(export_folder_path + "/" + export_totals_rows_removed_file_name_csv)
    #export_totals_rows_removed_file_path_pickle = Path(export_folder_path + "/" + export_totals_rows_removed_file_name_pickle)
    
    
    # Export file Name :: Totals Rows DF Export 
    #export_totals_rows_file_name_csv = "CDC_Natality_Birth_Data_" + str(year) + "_totals_rows.csv"
    #export_totals_rows_file_name_pickle = "CDC_Natality_Birth_Data_" + str(year) + "_totals_rows.pkl"

    # Export file Path :: Totals Rows DF Export 
    #export_totals_rows_file_path_csv = Path(export_folder_path + "/" + export_totals_rows_file_name_csv)
    #export_totals_rows_file_path_pickle = Path(export_folder_path + "/" + export_totals_rows_file_name_pickle)

    # Export CDC Data with Totals Rows Removed, leaving only the data 
    #totals_removed_df.to_csv(export_totals_rows_removed_file_path_csv, sep=",")

    #totals_removed_df.to_pickle(export_totals_rows_removed_file_path_pickle)

    # Export a Copy of the the removed total rows to a seperate file 
    #totals_rows_df.to_csv(export_totals_rows_file_path_csv, sep=",")
    
    #totals_rows_df.to_pickle(export_totals_rows_file_path_pickle)


    # Create Check Point File Name For Totals Row Removed
    checkpoint_export_file_name_total_rows_removed = "raw_CDC_Natality_Birth_Data_" + str(year) + "_totals_rows_removed"
    
    # Create Check Point For Totals Row Removed
    CheckPoint.create_checkpoint("raw_CDC_Natality_Birth_Data", checkpoint_export_file_name_total_rows_removed, totals_removed_df)

        
    # Create Check Point File Name For Totals Row 
    checkpoint_export_file_name_total_rows = "raw_CDC_Natality_Birth_Data_" + str(year) + "_totals_rows"

    # Create Check Point For Totals Row
    CheckPoint.create_checkpoint("raw_CDC_Natality_Birth_Data", checkpoint_export_file_name_total_rows, totals_rows_df)


    
    # Step 9 :: Delete Partial Totals Rows Removed CDC Data Dataframe to free up memory before looping again
    del totals_removed_df

    # Step 10 :: Delete Partial Totals Row CDC Data Dataframe to free up memory before looping again 
    del totals_rows_df

    del raw_cdc_natality_data_partial_df
```
